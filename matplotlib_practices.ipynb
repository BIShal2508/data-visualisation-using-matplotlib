```python
dt = 0.01
t = np.arange(0,10,dt)
nse = np.random.randn(len(t))
r = np.exp(-t/0.05)
```

```python
cnse = np.convolve(nse,r)*dt
cnse = cnse[:len(t)]
s = 0.1*np.sin(2*np.pi*t)+cnse
```

```python
plt.subplot(211)
plt.plot(t,s)
plt.subplot(212)
plt.psd(s,512,1/dt)
plt.show()
```

```python
x1 = np.linspace(0.0,5.0)
x2 = np.linspace(0.0,5.0)
y1 = np.cos(2*np.pi*x1)*np.exp(-x1)
y2 = np.cos(2*np.pi*x2)
```

```python
plt.subplot(211)
plt.plot(x1,y1,'o-')
plt.title('a table of 2 subplots')
plt.ylabel('Damped Oscillation')
plt.subplot(212)
plt.plot(x2,y2,'-')
plt.xlabel('time(s)')
plt.ylabel('Undamped')
plt.show()
```

```python
delta = 0.025
x = y = np.arange(-3,3,delta)
X,Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2-Y**2)
Z2 = np.exp(-(X-1)**2-(Y-1)**2)
Z = (Z1-Z2)*2
```

```python
mu = 100
sigma = 15
x = mu + sigma*np.random.randn(437)
num_bins = 50
fig,ax = plt.subplots()
n,bins,patches = ax.hist(x)
# best fit line (adding)
y =  ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))
ax.plot(bins,y,'--')
ax.set_xlabel('Smarts')
ax.set_ylabel('Probability density')
ax.set_title('Hstogram')
fig.tight_layout()
plt.show()
```

```python
def gridspec_graph():
    import matplotlib.gridspec as gridspec
    fig = plt.figure()
    spec = gridspec.GridSpec(ncols=3,nrows=3,figure=fig)
    g1 = fig.add_subplot(spec[0,:])
    g1.set_title('spec[0,:]')
    g2 = fig.add_subplot(spec[1,-1])
    g2.set_title('spec[1,-1]')
```

```python
def f(t):
    return np.exp(-t)*np.cos(2*np.pi*t)
def controlling_view_limits():
    t = np.arange(0.0,3.0,0.01)
    ax1 = plt.subplot(211)
    ax1.plot(t,f(t))
    ax2 = plt.subplot(212)
    ax2.plot(f(t),-t)
```
